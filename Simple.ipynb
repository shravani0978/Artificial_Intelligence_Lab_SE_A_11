```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
num = 5
print(f"The factorial of {num} is {factorial(num)}")

num = 0
print(f"The factorial of {num} is {factorial(num)}")

num = 7
print(f"The factorial of {num} is {factorial(num)}")
```

```python
def display_multiplication_table(number, limit=10):
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num_to_display = 7
display_multiplication_table(num_to_display)

print("\n") # Add a newline for better separation

num_to_display = 12
display_multiplication_table(num_to_display, limit=5)
```

```python
def check_odd_or_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 10
print(check_odd_or_even(num1))

num2 = 7
print(check_odd_or_even(num2))

num3 = 0
print(check_odd_or_even(num3))
```

```python
def is_palindrome(number):
    # Convert the number to a string
    num_str = str(number)
    # Compare the string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num1 = 121
print(is_palindrome(num1))

num2 = 123
print(is_palindrome(num2))

num3 = 1001
print(is_palindrome(num3))

num4 = -121 # Negative numbers are generally not considered palindromes in this context
print(is_palindrome(num4))
```

```python
# Simple Disease Expert System

print("Simple Disease Diagnosis System")

fever = input("Do you have fever? (yes/no): ")
cough = input("Do you have cough? (yes/no): ")
headache = input("Do you have headache? (yes/no): ")

if fever == "yes" and cough == "yes":
    print("Diagnosis: You may have Flu.")
elif fever == "yes" and headache == "yes":
    print("Diagnosis: You may have Viral Fever.")
elif cough == "yes" and headache == "yes":
    print("Diagnosis: You may have Common Cold.")
else:
    print("Diagnosis: Symptoms unclear. Please consult a doctor.")
```
